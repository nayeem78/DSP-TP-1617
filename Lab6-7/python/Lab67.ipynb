```python
##butterwoth order 3 for low pass


b, a = signal.butter(3, 0.1,'low') ##order 3
w, h = signal.freqs(b, a) ##continous, freqz for discrete
plt.plot(w, 20 * np.log10(abs(h)))
plt.xscale('log')
plt.title('Butterworth low pass filter')
plt.xlabel('Normalised frequency')
plt.ylabel('Gain')
plt.margins(0, 0.2)
#plt.grid(which='both', axis='both')
##plt.axvline(100, color='green') # cutoff frequency
plt.show()


##butterwoth order 3 for high pass
b, a = signal.butter(3, 0.1,'high', analog=True) ##order 3
w, h = signal.freqs(b, a) ##continous, freqz for discrete
plt.plot(w, 20 * np.log10(abs(h)))
plt.xscale('log')
plt.title('Butterworth highpass filter')
plt.xlabel('Normalised frequency')
plt.ylabel('gain')
plt.margins(0, 0.1)
# plt.grid(which='both', axis='both')
# plt.axvline(100, color='green') # cutoff frequency
plt.show()


##butterwoth order 3 for bandpass
b, a = signal.butter(3, [0.1,0.2],'bandpass', analog=True) ##order 3
w, h = signal.freqs(b, a) ##continous, freqz for discrete
plt.plot(w, 20 * np.log10(abs(h)))
plt.xscale('log')
plt.title('Butterworth banspass filter')
plt.xlabel('Normalised frequency')
plt.ylabel('gain')
plt.margins(0, 0.2)
# plt.grid(which='both', axis='both')
# plt.axvline(100, color='green') # cutoff frequency
plt.show()

##butterwoth order 3 for bandstop

b, a = signal.butter(3, [0.1,0.2],'bandstop', analog=True) ##order 3
w, h = signal.freqs(b, a) ##continous, freqz for discrete
plt.plot(w, np.log10(abs(h)))
plt.xscale('log')
plt.title('Butterworth bandstop filter')
plt.xlabel('Normalised frequency')
plt.ylabel('gain')
plt.margins(0, 0.2)
# plt.grid(which='both', axis='both')
# plt.axvline(100, color='green') # cutoff frequency
plt.show()
```

```python
def step(n=0, N=20):
    # Check that the parameter are valid
    if n > N - 1:
        raise ValueError('n should be inferior to N-1')
    
    # Create the array
    x = np.zeros((N, ))
    # Create the ramp
    for y in np.arange(n, N):
        x[y] = 1
    
    return x

N = 40
## Step function 
x = step(10,N) - step(30,N)
plt.figure
plt.stem(x)

```

```python
import numpy as np
import matplotlib.pyplot as plt
# from scipy import ndimage as ndi

# from skimage import img_as_float
# from matplotlib import cm
from skimage import rotate
# from scipy import rotate


from skimage import feature

# Generate noisy image of a square
im = np.zeros((128, 128))
im[32:-32, 32:-32] = 1

# im = barbara_im.rotate(im, 15, mode='constant')
im = barbara_im.gaussian_filter(im, 4)
im += 0.2 * np.random.random(im.shape)

# Compute the Canny filter for two values of sigma
edges1 = feature.canny(im)
edges2 = feature.canny(im, sigma=3)

# display results
fig, (ax1, ax2, ax3) = plt.subplots(nrows=1, ncols=3, figsize=(8, 3), sharex=True, sharey=True)

ax1.imshow(im, cmap=plt.cm.jet)
ax1.axis('off')
ax1.set_title('noisy image', fontsize=20)

ax2.imshow(edges1, cmap=plt.cm.gray)
ax2.axis('off')
ax2.set_title('Canny filter, $\sigma=1$', fontsize=20)

ax3.imshow(edges2, cmap=plt.cm.gray)
ax3.axis('off')
ax3.set_title('Canny filter, $\sigma=3$', fontsize=20)

fig.subplots_adjust(wspace=0.02, hspace=0.02, top=0.9,
                    bottom=0.02, left=0.02, right=0.98)

plt.show()

```
