```python
def step(n=0, N=20):
    # Check that the parameter are valid
    if n > N - 1:
        raise ValueError('n should be inferior to N-1')
    
    # Create the array
    s = np.zeros((N, ))
    # Create the ramp
    for y in np.arange(n, N):
        s[y] = 1
    
    return s

N = 40
## Step function 
x = step(10,N) - step(30,N)
plt.figure
plt.stem(x)

```

```python


def gradient_hor (Im):
    
    [nrow, ncol] = np.shape(Im)
    Im_s = np.zeros((nrow,ncol))
    Im_s_c = np.zeros((nrow,ncol))
    Im_s_n = np.zeros((nrow,ncol))
    Im_d_c = np.zeros((nrow,ncol))
    Im_d_n = np.zeros((nrow,ncol))


    #smoothing on columns
    alpha=1
    s=1 #as Ts is 1
    a=np.exp(-1)
    
    for i in range (2, ncol):
        Im_s_c[i,:] = Im[i,:] + a*(alpha - 1)*Im[i-1,:]  + 2*a*Im_s_c[i-1,:] - (a**2)*Im_s_c[i-2,:]
    for i in range (Im.shape[0]-3,-1, -1):
        Im_s_n[i,:] = a*(alpha + 1)* Im[i+1,:] - (a**2)*Im[i+2,:] + 2*a*Im_s_n[i+1,:] - (a**2)*Im_s_n[i+2,:]

    Im_smoothed = Im_s_c + Im_s_n

    
    #now derivative of the already smoothed columns
    for i in range (2, nrow):
        Im_d_c[:,i] = -s*alpha*a*Im_smoothed[:,i-1]   + 2*a*Im_d_c[:,i-1] - (a**2)*Im_d_c[:,i-2]
    for i in range (Im.shape[0]-3,-1, -1):
        Im_d_n[:,i] = s*alpha*a*Im_smoothed[:,i+1]   + 2*a*Im_d_n[:,i+1] - (a**2)*Im_d_n[:,i+2]

    Im_derivated = Im_d_c + Im_d_n
    
    return Im_derivated

x = gradient_hor (barbara_im)
plt.figure()
plt.title('horizontal smoothing')
io.imshow(x, cmap=plt.cm.gray)

#now the vertical

def gradient_vert (Im):
    [nrow, ncol] = np.shape(Im)
    Im_s = np.zeros((nrow,ncol))
    Im_s_c = np.zeros((nrow,ncol))
    Im_s_n = np.zeros((nrow,ncol))
    Im_d_c = np.zeros((nrow,ncol))
    Im_d_n = np.zeros((nrow,ncol))


    #smoothing on rows
    alpha=1
    s=1 #as Ts is 1
    a=np.exp(-1)
    
    for i in range (2, ncol):
        Im_s_c[:,i] = Im[:,i] + a*(alpha - 1)*Im[:,i-1]  + 2*a*Im_s_c[:,i-1] - (a**2)*Im_s_c[:,i-2]
    for i in range (Im.shape[0]-3,-1, -1):
        Im_s_n[:,i] = a*(alpha + 1)* Im[:,i+1] - (a**2)*Im[:,i+2] + 2*a*Im_s_n[:,i+1] - (a**2)*Im_s_n[:,i+2]

    Im_smoothed = Im_s_c + Im_s_n

    
    #now derivative of the already smoothed rows
    for i in range (2, nrow):
        Im_d_c[i,:] = -s*alpha*a*Im_smoothed[i-1,:]   + 2*a*Im_d_c[i-1,:] - (a**2)*Im_d_c[i-2,:]
        
    for i in range (Im.shape[0]-3,-1, -1):
        Im_d_n[i,:] = s*alpha*a*Im_smoothed[i+1,:]   + 2*a*Im_d_n[i+1,:] - (a**2)*Im_d_n[i+2,:]

    Im_derivated = Im_d_c + Im_d_n
    
    return Im_derivated


y = gradient_vert (barbara_im)
plt.figure()
plt.title('vertical smoothing')
io.imshow(y, cmap=plt.cm.gray)

```
